# Feb 30 days

```python
Feb = pd.read_csv('Feb_tran_id_expire.csv',index_col=['msno'])
```

```python
size=4e7 # 40 million
reader = pd.read_csv('user_logs.csv',chunksize=size,index_col=['msno'])
start_time = time.time()
for i in range(10):
    user_log_chunk=next(reader)
    
    
    user_log_chunk = user_log_chunk[(user_log_chunk.date<20170301)&(user_log_chunk.date>=20170101)]
    merged = pd.merge(user_log_chunk, Feb,left_index=True,right_index=True, how='inner')
    merged['date'] = pd.to_datetime(merged['date'].astype(str), format='%Y%m%d')
    merged['membership_expire_date'] = pd.to_datetime(merged['membership_expire_date'].astype(str), format='%Y%m%d')
    merged = merged[(merged['membership_expire_date'] > merged['date']) & 
      (merged['membership_expire_date'] <= pd.Timedelta(30, unit='D')+merged['date'])]
    
    if(i==0):
        result=merged
        print("Loop ",i,"took %s seconds" % (time.time() - start_time))
    else:
        result=result.append(merged)
        print("Loop ",i,"took %s seconds" % (time.time() - start_time))
        
    del(user_log_chunk)
    del(merged)
```

```python
result.to_csv("Feb_user_log.csv")
```

# Total Feb

```python
#writing as a function
def process_user_log(chunk):
    grouped_object=chunk.groupby(chunk.index,sort=False) # not sorting results in a minor speedup
    func = {'date':['max','count'], 
           'num_25':['sum'],'num_50':['sum'],
           'num_75':['sum'],'num_985':['sum'],
           'num_100':['sum'],'num_unq':['sum'],'total_secs':['sum'],'membership_expire_date':['min']}
    answer=grouped_object.agg(func)
    return answer
```

```python
size=4e7 # 40 million
reader = pd.read_csv('user_logs.csv',chunksize=size,index_col=['msno'])
start_time = time.time()
for i in range(10):
    
    user_log_chunk=next(reader)
    merged = pd.merge(user_log_chunk, Feb,left_index=True,right_index=True, how='inner')
    merged = merged[merged['membership_expire_date'] > merged['date']]
    
    
    if(i==0):
        result=process_user_log(merged)
        print("Loop ",i,"took %s seconds" % (time.time() - start_time))
    else:
        result=result.append(process_user_log(merged))
        print("Loop ",i,"took %s seconds" % (time.time() - start_time))
    del(user_log_chunk)
    del(merged)
```

```python
#result.columns.values
result.columns = ['date_max', 'date_count', 'num_25_sum', 'num_50_sum', 'num_75_sum',
       'num_985_sum', 'num_100_sum', 'num_unq_sum', 'total_secs_sum',
       'membership_expire_date']
```

```python
func_2 = {'date_max':['max'],'date_count':['sum'] ,
           'num_25_sum':['sum'],'num_50_sum':['sum'],
           'num_75_sum':['sum'],'num_985_sum':['sum'],
           'num_100_sum':['sum'],'num_unq_sum':['sum'],'total_secs_sum':['sum'],'membership_expire_date':['min']}
Feb_total=result.groupby(result.index).agg(func_2)
Feb_total.columns = Feb_total.columns.get_level_values(0)
Feb_total
```

```python
Feb_total['date_max'] = pd.to_datetime(Feb_total['date_max'].astype(str), format='%Y%m%d')
Feb_total['membership_expire_date'] = pd.to_datetime(Feb_total['membership_expire_date'].astype(str), format='%Y%m%d')
Feb_total['last_day']  = Feb_total['membership_expire_date'] - Feb_total['date_max']
```

```python
Total = (Feb_total['num_25_sum']+Feb_total['num_50_sum']
         +Feb_total['num_75_sum']+Feb_total['num_985_sum']
         +Feb_total['num_100_sum'])

Feb_total['percentage_length'] = (Feb_total['num_25_sum']*0.25+Feb_total['num_50_sum']*0.5
                                          +Feb_total['num_75_sum']*0.75+Feb_total['num_985_sum']*0.985
                                          +Feb_total['num_100_sum'])/Total

Feb_total['percent_whole_song'] = (Feb_total['num_985_sum']+Feb_total['num_100_sum'])/Total
Feb_total['percent_lower25_song'] = Feb_total['num_25_sum']/Total
Feb_total['percent_unique'] = Feb_total['num_unq_sum']/Total
```

```python
def process_user_log_2(chunk):
    grouped_object=chunk.groupby(chunk.index,sort=False) # not sorting results in a minor speedup
    func = {'date':'count', 
           'sec_dif_square':'sum'}
    answer=grouped_object.agg(func)
    return answer
```

```python
#std
mean = pd.DataFrame(Feb_total['total_secs_sum']/Feb_total['date_count'])
mean.columns = ['mean_sec']
size=4e7 # 40 million
reader = pd.read_csv('user_logs.csv',chunksize=size,index_col=['msno'])
start_time = time.time()
for i in range(10):
    
    user_log_chunk=next(reader)
    merged = pd.merge(user_log_chunk, Feb,left_index=True,right_index=True, how='inner')
    merged = merged[merged['membership_expire_date'] > merged['date']]
    merged = pd.merge(merged,mean,left_index=True,right_on = 'msno',how = 'inner')
    merged['sec_dif_square'] = np.square(merged['total_secs']-merged['mean_sec'])
    
    if(i==0):
        result=process_user_log_2(merged)
        print("Loop ",i,"took %s seconds" % (time.time() - start_time))
    else:
        result=result.append(process_user_log_2(merged))
        print("Loop ",i,"took %s seconds" % (time.time() - start_time))
    del(user_log_chunk)
    del(merged)
```

```python
std_sec = result.groupby('msno').agg({'date':'sum','sec_dif_square':'sum'})
std_sec = pd.DataFrame(std_sec.sec_dif_square/std_sec.date)
std_sec.columns = ['std_sec']
```

```python
Feb_total = pd.merge(Feb_total,std_sec,left_index=True,right_index=True, how='inner')
Feb_total['last_day'] = Feb_total.last_day.dt.days.astype('int16')
Feb_total
```

```python
Feb_total.to_csv("Feb_total.csv")
```
